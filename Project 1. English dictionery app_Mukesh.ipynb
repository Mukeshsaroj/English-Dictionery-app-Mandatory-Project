```python
'''Main Menu

Add a new word
Find the meaning
Update a word
Exit
        Enter Choice:

If item 1 is selected, it should prompt for a word and then accept its meaning. 
This pair of word and its meaning should be stored in a file called ‘words.txt’.
When item 2 is selected, it should prompt for a word, search that word in the file ‘words.txt’ 
and return the meaning if it is found. If the meaning is not found, it should display an appropriate message. 
Similarly, when choice 3 is entered, it should accept a word and update its meaning. In all the above three cases, 
the prompt should return to the main menu. The words and their meanings should be stored as a dictionary in words.txt, 
using the concept of serialization. If choice 4 is selected, a graceful exit should be performed.'''


import json
choice=1
while(choice!=4):
    choice=int(input('''Main Menu
    Press 1... to Add new word
    Press 2... to Search word
    Press 3...to Update the word
    Press 4... to Exit
Enter the choice : '''))
    print("Your choice is : ",choice)
    if (choice==2):
        word=input("Enter the word you want to search :").title()
        with open('words.txt','r') as f:
            a=json.load(f)
            for i in a:
                if (word==i):
                    print("Meaning of : ",word,"is :",a[i])
                    break
            else:
                print("No word in dictionery you can try adding")
    elif(choice==1):
        temp={}
        temp1={}
        word=input("Enter the new word : ").title()
        with open("words.txt",'r')as f:
            a=json.load(f)
            for i in a:
                if(word==i):
                    print("Word already in dictionery try updating")
                    break
            else:
                meaning=input("Enter the meaning :").title()
                temp[word]=meaning
                for i in a,temp:
                    temp1.update(i)
        with open("words.txt",'w')as m:
            json.dump(temp1,m)
    elif (choice==4):
        print("Thank You")
    elif(choice==3):
        temp={}
        temp1={}
        word=input("Enter the word to update : ").title()
        with open("words.txt",'r')as f:
            a=json.load(f)
            for i in a:
                if(word==i):
                    meaning=input("Enter the meaning : ").title()
                    temp[word]=meaning
                    break
            else:
                print(word,"is not in list try adding")
        for i in a,temp:
            temp1.update(i)
        with open("words.txt",'w')as f:
            json.dump(temp1,f)
    else:
        print("please choose between above options....")
        
```
